```python
data.dropna(subset=['tags','cooking_method'], how='any', inplace=True)

data = data.reset_index(drop=True)
```

```python
data.isnull().sum() 
```

```python
data.shape
```

```python
target = data.tags.unique()
```

```python
labels = list()
```

```python
for item in target:
    temp = item.split(',')
    for x in temp:
        if x.strip() not in labels:
            labels.append(x.strip()) 
    else:
            continue
```

```python
len(labels)
```

```python
 labels
```

```python
for label in labels:
    data[label]=0
```

```python
data.head()
```

```python
data.shape
```

```python
for index, row in data.iterrows():
    for label in labels:
        if label in row['tags']:
            data.at[index, label] = 1
```

```python
import nltk
nltk.download()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
from sklearn.multiclass import OneVsRestClassifier
from sklearn.pipeline import Pipeline
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
train, test = train_test_split(data, random_state=42, test_size=0.33, shuffle=True)
```

```python
X_train = train.cooking_method
X_test = test.cooking_method
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LogReg_pipeline = Pipeline([
                ('tfidf', TfidfVectorizer(stop_words=stop_words)),
                ('clf', OneVsRestClassifier(LogisticRegression(solver='sag'), n_jobs=1)),
            ])
for label in labels:
    print('... Processing {}'.format(label))
    LogReg_pipeline.fit(X_train, train[label])
    
    # compute the testing accuracy
    prediction = LogReg_pipeline.predict(X_test)
    print('Test accuracy is {}'.format(accuracy_score(test[label], prediction)))
```
